6.Write a code to perform basic string compression using the counts of repeated characters .

Ans-

   def compress_string(s):  
  if not s:  
       return ""  

   compressed = []  
   count = 1  

  for i in range(1, len(s)):  
        if s[i] == s[i - 1]:  
            count += 1  
        else:  
            compressed.append(s[i - 1])  
            compressed.append(str(count))  
            count = 1  

    # Append the last character and its count  
   compressed.append(s[-1])  
   compressed.append(str(count))  

   compressed_string = ''.join(compressed)  
    
    # Return the compressed string only if it's shorter than the original  
   return compressed_string if len(compressed_string) < len(s) else s  

# Example usage  
input_string = input("Enter a string to compress: ")  
compressed_string = compress_string(input_string)  

print(f"Compressed string: {compressed_string}")  

Explanation:
The function compress_string takes a string s as input.
It initializes a compressed list and a count variable to keep track of the number of consecutive characters.
It iterates through the string, comparing each character with the previous one:
If they are the same, it increments the count.
If they are different, it appends the previous character and its count to the compressed list, then resets the count.
After the loop, it appends the last character and its count.
Finally, it converts the compressed list back to a string and returns it, but only if the compressed version is shorter than the original string.

14.Write a code to reverse a  list in-place without using any
 built-in reverse functions.


Ans- We can reverse a list in-place without using any built-in reverse functions by swapping elements from the start and end of the list until we reach the middle. Here’s a simple implementation in Python:


def reverse_list_in_place(input_list):  
     left = 0  
     right = len(input_list) - 1  

     while left < right:  
        # Swap the elements at left and right indices  
        input_list[left], input_list[right] = input_list[right], input_list[left]  
        
        # Move towards the middle  
        left += 1  
        right -= 1  

# Example usage  
my_list = [1, 2, 3, 4, 5]  
reverse_list_in_place(my_list)  

print("Reversed List:", my_list)  

Explanation:
Two Pointers: The left pointer starts at the beginning of the list, and the right pointer starts at the end.
Swapping: While the left pointer is less than the right pointer, the elements at these positions are swapped.
Move Pointers: After each swap, the left pointer increments, and the right pointer decrements, moving towards the middle of the list.
In-Place: This method modifies the list directly without creating a new list, which counts as an in-place operation.
Example Output:
For the list [1, 2, 3, 4, 5], the output will be:

Reversed List: [5, 4, 3, 2, 1]

20.Write a code to Shuffle a  given list randomly
 without using any
 built-in shuffle functions.


Ans-  We can shuffle a given list randomly without using built-in shuffle functions by implementing the Fisher-Yates shuffle (also known as the Knuth shuffle) algorithm. This algorithm efficiently shuffles the list in place. Here’s how we can do it:


import random  

def shuffle_list(arr):  
    # Loop through the list in reverse order  
     for i in range(len(arr) - 1, 0, -1):  
        # Generate a random index from 0 to i  
        j = random.randint(0, i)  
        # Swap elements at index i and j  
        arr[i], arr[j] = arr[j], arr[i]  

# Example usage  
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]  
print("Original list:", my_list)  

shuffle_list(my_list)  
print("Shuffled list:", my_list)  

Explanation:
Looping: The loop iterates through the list from the last element to the first (excluding the first element).

Random Index: For each element at index i, a random index j is generated using random.randint(0, i). This ensures that the swapping happens within the bounds of the array.

Swapping: The elements at indices i and j are swapped. This effectively randomizes the list.

In-Place: The original list is shuffled in place without the need for additional list storage.

Example Output:

For the original list [1, 2, 3, 4, 5, 6, 7, 8, 9], the output after shuffling could be:


Original list: [1, 2, 3, 4, 5, 6, 7, 8, 9]  
Shuffled list: [3, 7, 1, 5, 6, 2, 8, 4, 9]  

Every time we run the program, we may get a different shuffled order, demonstrating randomness in the output.

Common elements: (4, 5)
